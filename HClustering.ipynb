```r
# read data
df <- read.csv("hclustmovies.csv")
```

```r
head(df)
```

```r
df <- df[, -1]
```

```r
head(df)
```

```r
install.packages("proxy")
library(proxy)
```

```r
# Considering Revenue, Runtime and Budget columns for clustering
temp1 <- df[1:200, c('revenue', 'budget', 'runtime')]
```

```r
# Perform hierarchical clustering with cosine distance
cos_sim_matrix <- as.matrix(dist(temp1, method = "cosine"))
```

```r
# Convert similarity matrix to distance matrix
cos_dist_matrix <- as.dist(cos_sim_matrix)
```

```r
# Using Ward's method for clustering
hc <- hclust(cos_dist_matrix,method = 'ward.D')
```

```r
# Plotting dendrogram for 2 clusters

plot(hc, main = "Revenue, Budget and Runtime 2-cluster Dendrogram")
rect.hclust(hc,k=2,border=2:5)
```

```r
# Plotting dendrogram for 3 clusters
plot(hc, main = "Revenue, Budget and Runtime 3-cluster Dendrogram")
rect.hclust(hc,k=3,border=2:5)
```

```r
# Plotting dendrogram for 4 clusters
plot(hc, main = "Revenue, Budget and Runtime 4-cluster Dendrogram")
rect.hclust(hc,k=4,border=2:5)
```
